#                Label Encoder

```python
df2.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df2.sex=le.fit_transform(df2.sex)
df2.smoker=le.fit_transform(df2.smoker)
df2.region=le.fit_transform(df2.region)
df2.head()
```

#            Label Encoder using Loop

```python
df3.head()
```

```python
df3.columns
```

```python
from pandas.core.dtypes.common import is_numeric_dtype
```

```python
for col in df3.columns:
    if is_numeric_dtype(df3[col]):
        continue
    else:
        df3[col]=le.fit_transform(df3[col])
```

```python
df3.head()
```

#            One Hot Encoder

```python
df4.head()
```

```python
for col in df4.columns:
    if is_numeric_dtype(df4[col]):
        continue
    else:
        dum = pd.get_dummies(df4[col],prefix=col,drop_first=True)
        df4=pd.concat([df4,dum],axis=1)
        df4=df4.drop(col,axis=1)
```

```python
df4.head()
```

# Ordinal

```python
df5.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
var=df5.sex.unique()
ordinal=OrdinalEncoder(categories=[var])
encoded=ordinal.fit_transform(df5[['sex']])   # Data in Arra
en=pd.DataFrame(encoded,columns=['SEX'])  # convert to the Data Frame
df5=df5.drop('sex',axis=1)   # Drop sex column
df5=pd.concat([df5,en],axis=1)   # Concanate en column
```

```python
var=df5.smoker.unique()
ordinal=OrdinalEncoder(categories=[var])
encoded=ordinal.fit_transform(df5[['smoker']])   # Data in Arra
en=pd.DataFrame(encoded,columns=['SMOKER'])  # convert to the Data Frame
df5=df5.drop('smoker',axis=1)   # Drop smoker column
df5=pd.concat([df5,en],axis=1)   # Concanate en column
```

```python
var=df5.region.unique()
ordinal=OrdinalEncoder(categories=[var])
encoded=ordinal.fit_transform(df5[['region']])   # Data in Arra
en=pd.DataFrame(encoded,columns=['REGION'])  # convert to the Data Frame
df5=df5.drop('region',axis=1)   # Drop region column
df5=pd.concat([df5,en],axis=1)   # Concanate en column
```
